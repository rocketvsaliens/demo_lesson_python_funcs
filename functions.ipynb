Функция в Python - это как маленький помощник, который делает работу за тебя. Представь, что ты хочешь приготовить пирог, но у тебя нет времени на все этапы. Ты можешь попросить функцию "смешать ингредиенты" или "проверить готовность". Она сделает это за тебя, а ты потом просто соединишь все вместе и получишь готовый пирог. 

Это как если бы ты научил маленького робота делать что-то, а потом мог бы его попросить об этом в любое время.

Давай попробуем сделать прототип такого робота на примере функции **cook()**. Сначала напишем обычную программу для приготовления пирога.

```python
# Задаём параметры для пирога

pie_title = 'Пирог с яблоками'  # Это название нашего пирога

ingridients = ['мука', 'яйца', 'соль', 'молоко', 'яблоки']  # Список ингридиентов

cooking_method = 'Запекаем в духовке'  # Метод приготовления

# Готовим наш "пирог"
print(f'Для приготовления блюда {pie_title} смешиваем:')
for ingridient in ingridients:
    print(ingridient)
print()
print('Готовим пирог: ' + cooking_method)
print()
print('Проверяем готовность...')
print()
print(f'Ура, {pie_title} готов!')
```

### Но что будет, если мы захотим приготовить не пирог, а салат или яичницу (•ิ_•ิ)?

Нам придётся переписывать весь код программы (_но программисты терпеть не могут повторений участков кода_) или мы можем написать **_функцию_**!

```python
def cook(dish_title, list_of_ingridients, cooking_method):
    print(f'Для приготовления блюда {dish_title} смешиваем:')
    for ingridient in list_of_ingridients:
        print(ingridient)
    print()
    print('Готовим блюдо: ' + cooking_method)
    print()
    print('Проверяем готовность...')
    print()
    print(f'Ура, наше блюдо {dish_title} готово!')
```

Мы выполнили ячейку, но ничего не произошло... Почему?
### Просто надо добавить немного магии Python! ٩(◕‿◕｡)۶

```python
# Задаём параметры для блюда

dish_title = 'Салат из огурцов и помидоров'  # Это название нашего пирога

list_of_ingridients = ['огурцы', 'помидоры', 'листья салата', 'соль']  # Список ингридиентов

cooking_method = 'Заправить маслом'  # Метод приготовления
```

Мы выполнили и эту ячейку, но снова ничего не произошло... ╮(￣_￣)╭

```python
# А вот и магия...

cook(dish_title, list_of_ingridients, cooking_method)
```

### Как видишь, приготовили блюдо, уложившись всего в одну строчку (✯◡✯)
Ну почти...

А можно и так:

```python
cook(dish_title='Яичница', list_of_ingridients=['3 куриных яйца', 'масло'], cooking_method='жарим на сковородке')
```

При помощи нашего маленького робота функции cook() мы можем теперь приготовить практически любое блюдо за одну строчку!

Хорошо, максимум за 4 строчки кода. Но ведь изначально их было куда больше!

В Python функция -- это набор инструкций, которые выполняются вместе. Ты можешь создать свою функцию и назвать её как угодно, а потом использовать её в своём коде.

**Но вернёмся к самой популярной у начинающих программистов функции print()**

"Пустая" функция print() печатает пустую строку.

Чтобы она напечатала нам то, что мы хотим, мы должны сказать, что именно она должна напечатать. Или, как говорят программисты:
### передать аргументы в функцию   (ノ ˘_˘)ノ　ζ|||ζ　ζ|||ζ　ζ|||ζ

```python
print('Привет, хакеры!')  # print -- это имя функции, по которому мы её вызываем. В скобках после имени находятся аргументы функции.
```

```python
print('Привет', 'хакеры!')
print('Привет ' + 'хакеры!') 
```

Как видишь, во всех трёх случаях мы получили одно и то же предложение.

Это значит, что функция **print()** может принимать разное число агрументов и обрабатывать их по-разному.

Например, она может принимать цифры и выполнять с ними какие-то вычисления.

```python
print(1 + 1 * 7)
```

```python
print((1 + 1) * 7)  # вторые скобки относятся уже к примеру для вычисления, а не к функции
```

### Что обязательно надо знать про функции в Python?
**Функция** -- это алгоритм, составленный на языке программирования и имеющий уникальное имя.

**Когда мы создаём собственную функцию, то всегда помним, что:**
1. Каждая функция начинается с волшебного слова **def** (от английского "define" -- объявлять).
2. После слова **def** мы пишем название функции. Почему оно должно быть уникальным в пределах одной программы -- смотри картинку внизу.
3. После имени функции мы пишем две скобки **()**. В них мы можем передать аргументы: то, с чем наша функция будет работать.
+ **(!)**_Даже если функции не нужно никаких аргументов -- такое тоже бывает -- скобки ей всё равно нужны. Программисты говорят, "функция не принимает никаких аргументов"_
+ _Не забудь про двоеточие после скобок, дальше мы будем перечислять инструкции. Вот так: первая, вторая, третья;_
4. Все инструкции ниже названия функции должны отстоять от левого края миниум на 4 пробела или один Tab. Программисты называют такие строчки _"телом функции"_.

![Функции в Python](img/unique_names_for_funcs.png)

Если бы у всех функций были одинаковые имена, то программа не смогла бы понять, какую функцию нужно выполнить. Представь, что у тебя есть две книги, которые называются одинаково. Как ты будешь знать, какую книгу тебе нужно взять? Вот и компьютеру нужно знать, какую функцию ему нужно выполнить, поэтому у каждой функции должно быть уникальное имя.

Когда ты установил Python, вместе с ним у тебя появилось много хороших и полезных функций.

Вот некоторые:
- **len()** -- вычисляет длину чего бы то ни было, что ты передал ей в качестве аргумента. Например, строки или списка.
- **sum()** -- складывает все цифры в списке, кортеже или другом итерируемом объекте.
- **max()** -- определяет максимальное значение среди двух или в списке (и любом другом интерируемом объекте).
- **min()** -- вычисляет минимальное значение.
- **input()** -- принимает информацию от пользователя.
- **type()** -- показывает тип тех данных, которые ты передал ей в качестве аргумента.

```python
# Запусти код ниже и поэкспериментируй с ним. Помни, что для того, чтобы ты увидел результат работы функции на экране, необходимо вызвать функцию print()

hello = 'Hello, World!'

print(len(hello))
```

```python
name = input('Напиши мне своё имя: ')

print(f'Привет, {name}!')

print(type(name))  # Обрати внимание, что даже если input принимает цифры в качестве аргумента, она преобразует их в строку
```

```python
# Напоследок поэкспериментируем с цифрами

lst = [1, 3, 547, 165412, 55, 67, 999964, -4, 7, 0]

print(f'Наименьшее число = {min(lst)}')
print(f'Наибольшее число = {max(lst)}')
print(f'Сумма всех элементов в списке = {sum(lst)}')
```

```python
# А теперь объединим все наши вычисления и добавим кое-что ещё
def do_calc_with_numbers(lst):
    print(f'Наименьшее число = {min(lst)}')
    print(f'Наибольшее число = {max(lst)}')
    print(f'Сумма всех элементов в списке = {sum(lst)}')
    print(f'Среднее арифметическое списка = {sum(lst) / len(lst)}')  # правильно, мы добавили среднее арифметическое!
    print('-------------')


# Поэкспериментируй со списком в агрументе функции, поменяй значения на строковые.
# Попробуй запустить функцию do_calc_with_numbers() с пустым списком.
# Или вообще без аргументов. 
# Не бойся ошибок, мы учимся на них ;)

# Вызываем нашу функцию по вычислению всего столько раз, сколько захотим!
do_calc_with_numbers(lst=[1, 4, 8, 16])

do_calc_with_numbers(lst=[54, 78, 30, 46, 1000000000])

do_calc_with_numbers(lst=[0, 0, 0, 0, 0])
```
